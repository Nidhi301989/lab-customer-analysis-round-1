```python
"""1 Read the three files into python as dataframes"""

import pandas as pd

# Define the file paths using forward slashes
file1_path = "C:/Users/ashis/Downloads/file1.csv"
file2_path = "C:/Users/ashis/Downloads/file2.csv"
file3_path = "C:/Users/ashis/Downloads/file3.csv"

# Read the files into pandas dataframes
df1 = pd.read_csv(file1_path)
df2 = pd.read_csv(file2_path)
df3 = pd.read_csv(file3_path)

# Display the first few rows of each dataframe
print(df1.head())
print(df2.head())
print(df3.head())

# Display the shape of each dataframe
print("Shape of df1:", df1.shape)
print("Shape of df2:", df2.shape)
print("Shape of df3:", df3.shape)


# Standardize header names
df1.columns = df1.columns.str.lower().str.replace(' ', '_')
df2.columns = df2.columns.str.lower().str.replace(' ', '_')
df3.columns = df3.columns.str.lower().str.replace(' ', '_')

# Display the standardized column names
print("Standardized columns for df1:", df1.columns.tolist())
print("Standardized columns for df2:", df2.columns.tolist())
print("Standardized columns for df3:", df3.columns.tolist())

desired_order = [ 'st', 'customer_lifetime_value',  
                 'monthly_premium_auto', 'education', 'number_of_open_complaints','gender', 
                 'policy_type', 'vehicle_class', 'income','customer',
                 'total_claim_amount']

    
df1 = df1[desired_order]
df2 = df2[desired_order]
""""df3 = df3[desired_order]"""

# Display the shape of each dataframe and the first few rows to verify the column order
print("Shape of df1:", df1.shape)
display(df1.head())
print("\nShape of df2:", df2.shape)
display(df2.head())
# print("\nShape of df3:", df3.shape)
display(df3.head())


combined_df = pd.concat([df1, df2, df3], ignore_index=True)

# Display the shape of the combined dataframe and the first few rows
print("Shape of combined_df:", combined_df.shape)
display(combined_df.head())

numerical_columns = combined_df.select_dtypes(include=['int64', 'float64']).columns.tolist()

# Display the numerical columns
display("Numerical columns:", numerical_columns)

categorical_columns = combined_df.select_dtypes(include=['object']).columns.tolist()

print("Categorical columns:", categorical_columns)

# Understand the meaning of all columns
description = combined_df.describe()

print(description)

# Drop 'education' and 'number_of_open_complaints' columns
combined_df = combined_df.drop(['education', 'number_of_open_complaints'], axis=1)

# Display the shape of the updated dataframe and the first few rows
print("Shape of combined_df after dropping columns:", combined_df.shape)
display(combined_df.head())


combined_df['customer_lifetime_value'] = combined_df['customer_lifetime_value'].str.rstrip('%').astype(float) * 100

# Display the updated DataFrame
print("\nDataFrame after correction:")
display(combined_df)

# Check the data types after conversion
print("\nData types of columns:")
print(combined_df.dtypes)

duplicate_rows = combined_df[combined_df.duplicated()]

if not duplicate_rows.empty:
    print("\nDuplicate rows:")
    display(duplicate_rows)
    
    # Remove duplicate rows
    combined_df = combined_df.drop_duplicates()

    # Display the DataFrame after removing duplicates
    print("\nDataFrame after removing duplicates:")
    display(combined_df)
else:
    print("\nNo duplicate rows found.")

# Check the shape of the DataFrame after removing duplicates
print("\nShape of DataFrame after removing duplicates:", combined_df.shape)

# Filter customers with income <= 0
filtered_df = combined_df[combined_df['income'] <= 0]

# Display the filtered DataFrame
print("\nFiltered DataFrame for customers with income <= 0:")
display(filtered_df)
```
